```python
import pandas as pd
import numpy as np
import math
```

```python
# embedding the feature vector of genes in benchmark dataset
df1 = pd.read_csv('data/cancer_weight.csv')
df1['source_target'] = df1['source'].astype(str)+'_'+df1['target'].astype(str)
del df1['source']
del df1['target']
df2 = pd.read_csv('data/cancer_weight.csv')
df2[['source', 'target']] = df2[['target', 'source']]
df2['source_target'] = df2['source'].astype(str)+'_'+df2['target'].astype(str)
del df2['source']
del df2['target']
df3 = df1.append(df2)
```

```python
df4 = pd.read_csv('data/MT_label.csv',header=0)
#df4
a = df4['GeneID'].tolist()
df5 = pd.DataFrame(0, index=a, columns=a)
df6 = df5.stack().reset_index()
df6.rename(columns={'level_0':'source','level_1':'target','0':'count'}, inplace = True)
df6['source_target'] = df6['source'].astype(str)+'_'+df6['target'].astype(str)
del df6['source']
del df6['target']
b = df6[0]
df6.drop(0,axis=1)
df6.insert(0,'weight',b)
del df6[0]
df6
```

```python
df7 = pd.merge(df3,df6,on='source_target')
del df7['weight_y']
df7.rename(columns={'weight_x':'weight'},inplace=True)
df7
```

```python
df7.to_csv('data/MT_weightedge.csv',index=0)
```

```python
#calculate PiPj
df8 = pd.read_csv('data/MT_label.csv',header=0)
#df8.rename(columns={'GeneID_x':'GeneID'},inplace=True)
df9 = pd.read_csv('data/singlegene_paper.csv',header=0)
df9.columns = ['GeneID','count']
#df9
allsinglegene_paper = df9['count'].sum()
```

```python
for i in df8.index:
    for j in df9.index:
        if(df9.at[j,'GeneID']==df8.at[i,'GeneID']):
            df8.at[i,'count'] = df9.at[j,'count']
            
for i in df8.index:
    df8.at[i,'P'] = df8.at[i,'count']/allsinglegene_paper
```

```python
df8.fillna(0,inplace=True)
#df8
df8.to_csv('data/MT_count_p.csv',index=0)
```

```python
a = {'P':df8['P']}
df9 = pd.DataFrame(data=a)
#df9
df10 = np.dot(df9,df9.T)
np.savetxt('data/MT_PiPj_sorted.csv', df10, delimiter = ',')
aa = df8['GeneID'].values
```

```python
df11 = pd.read_csv('data/MT_PiPj_sorted.csv',header=None)
df11.index = aa
df11.columns = aa
df11.to_csv('data/MT_PiPj_sorted_indexed.csv')
```

```python
#calculate Pij*pij/PiPj
Pij = pd.read_csv('data/MT_Pij.csv',header=0,index_col=0)
PiPj = pd.read_csv('data/MT_PiPj_sorted_indexed.csv',header=0,index_col=0)
df20 = Pij/PiPj 
df20.replace(0, np.nan, inplace=True)
```

```python
#calculate lg
list_a = df20._stat_axis.values.tolist() # row name
list_b = df20.columns.values.tolist()    # col name
for m in list_a:
    for n in list_b:
         df20.at[m,n] = math.log(df20.at[m,n],10)
```

```python
df21 = df20*Pij
df22 = df21.replace(np.inf,0)
df22.fillna(0,inplace=True)
df22
```

```python
df22.to_csv('data/MT_lg_label.csv')
```
